```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

data = pd.read_csv("./abalone19-5-1tra(1).csv")
# 计算不平衡比
balance_values = (data['class'].value_counts() / data["class"].count()).values
print("不平衡比是:", balance_values[0]/balance_values[1])
print("数据的形状是:",data.shape)
display(data.head())
data_class = data.pop('class')
len_data = len(data)
data.loc[len_data]  = data.mean()
display(data.tail())
```

```python
len(data)
```

```python
from scipy.spatial.distance import pdist, squareform
# 生成距离矩阵
distance = list()
data_inv = np.linalg.inv(np.cov(data.T))
# 打印逆协方差矩阵
print(data_inv)
print(data_inv.shape)

# 计算距离
distance = list()
for i in range(0,3340):
    distance.append(pdist(data.loc[[i,3339]], 'mahalanobis', VI=data_inv).tolist())
    if i%1000==0:
        print(i)
```

```python
# 添加距离
data['distance'] = np.array(distance)
# 添加class
data['class'] = data_class
# 按马氏距离排序
data.sort_values("distance",inplace=True)
# 删除平均值
data = data[1:-1]
# 初始化所有组
data['group'] = 0
# 重置索引
data = data.reset_index(drop=True)
# 按K分组
K=10
K1 = len(data)/K
data['group'] =  data.index.values // K1
```

```python
data.groupby('group').count()
```

```python
data2 = data.groupby('group')['class'].agg([min,max])
'''
# 正域
data2[data2['max'] == 1].index
# 负域
data2[data2['min'] == 2].index
# 不确定域
data2[ (data2['min'] == 1) & (data2['max'] == 2 )].index
'''
# 去掉负域的数据  取反 用data_choose来表示后来色数据
data_choose = data[~data['group'].isin( (data2[data2['min'] == 2].index).tolist() )]
print('有',len((data2[data2['min'] == 2].index).tolist()),'组被去除了')
```

```python
data_choose
```

```python
data_target = data_choose.pop('class')
```

## 使用决策树进行训练

```python
from sklearn.tree import DecisionTreeClassifier
from imblearn.metrics import geometric_mean_score
```

```python
estimator = DecisionTreeClassifier()
estimator.fit(data_choose,data_target)
y_pred = estimator.predict(data_choose[:100])
```

```python
geometric_mean_score(data_target[:100],y_pred)
```
